```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_text(text, chunk_size=400, chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  chunks = text_splitter.split_text(text)
  return chunks
```

```python
import pinecone

# connect to pinecone environment
pinecone.init(
    api_key="45682637-3050-4490-b886-884db0f41035",
    environment="us-west1-gcp-free"
)
```

```python
import torch
from sentence_transformers import SentenceTransformer

# set device to GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
# load the retriever model from huggingface model hub
retriever = SentenceTransformer(model_1, device=device)
retriever
```

```python
index_name = "extractive-question-answering"

# check if the extractive-question-answering index exists
if index_name not in pinecone.list_indexes():
    # create the index if it does not exist
    pinecone.create_index(
        index_name,
        dimension=retriever.get_sentence_embedding_dimension(),
        metric="cosine"
    )

# connect to extractive-question-answering index we created
index = pinecone.Index(index_name)
```

```python
text_chunks = split_text(cindrella_story)
```

```python
def embed_and_store(chunks):
  embeds = retriever.encode(chunks).tolist()
  metadata = [{'id': i, 'context': chunk} for i, chunk in enumerate(chunks)]
  ids = [f"{idx}" for idx in range(len(chunks))]
  to_upsert = list(zip(ids, embeds, metadata))
  # print(to_upsert)
  _ = index.upsert(vectors=to_upsert)
```

```python
embed_and_store(text_chunks)
```

```python
# gets context passages from the pinecone index
def get_context(question, top_k):
    # generate embeddings for the question
    xq = retriever.encode([question]).tolist()
    # search pinecone index for context passage with the answer
    xc = index.query(xq, top_k=top_k, include_metadata=True)
    # extract the context passage from pinecone search result
    c = [x["metadata"]["context"] for x in xc["matches"]]
    return c
```

```python
context = get_context(question, 1)
```

```python
context
```

```python

# extracts answer from the context passage
def extract_answer(question, context):
    results = []
    for c in context:
        # feed the reader the question and contexts to extract answers
        answer = qa(question=question, context=c)
        # add the context to answer dict for printing both together
        answer["context"] = c
        results.append(answer)
    # sort the result based on the score from reader model
    sorted_results = sorted(results, key=lambda x: x["score"], reverse=True)
    return sorted_results

```

```python
answer = extract_answer(question, context)
```
